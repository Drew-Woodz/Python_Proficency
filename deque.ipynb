```python
from collections import deque
import random

class RollingMaximum:
    def __init__(self, k):
        self.deque = deque(maxlen=k)  # Window of k items
        self.max_deque = deque()      # Indices of potential max values (decreasing order)
        self.k = k

    def find_max(self) -> int:
        # Remove indices no longer in the current window
        while self.max_deque and self.max_deque[0] < len(self.deque) - self.k:
            self.max_deque.popleft()
        # Return max value or None if no valid max
        return self.deque[self.max_deque[0]] if self.max_deque else None

    def cycle(self):
        # Add new random item
        new_item = random.randint(0, 9)
        self.deque.append(new_item)
        # Remove indices of smaller values from max_deque
        while self.max_deque and self.deque[self.max_deque[-1]] < new_item:
            self.max_deque.pop()
        # Add new item's index
        self.max_deque.append(len(self.deque) - 1)
        # Print results
        print(self.deque)
        print("Total:", sum(self.deque))
        print("Max value:", self.find_max() if len(self.deque) == self.k else "Filling window")

# Test it
k = 10
r = RollingMaximum(k)
for i in range(20):
    r.cycle()
```
